# Visualize data from MongoDB using Dash

```python
# Callback to update the bar chart-sales
@app.callback(
    Output("sales-bar-chart", "figure"),
    Input("category-selector", "value")
)
def update_bar_chart(category):
    # Group data by the selected category
    
    #category="store location"
    res=sales_weather.aggregate([
        {"$group": {"_id": f"${category}", "total_sales": {"$sum": "$sales_amount"}}},
        {"$sort": {"total_sales": -1}}  # Optional: Sort by total sales
        ])
    filtered_data = pd.DataFrame(res).reset_index();
    print(filtered_data)
    # Create bar chart
    fig = px.bar(filtered_data, x="_id", y="total_sales", title=f"Sales Amount by {category}", 
             labels={"total_sales": "Sales Amount", "_id": category},
             color_discrete_sequence=["skyblue"])

    # Show the figure
    #fig.show()
    return fig

@app.callback(
    Output("order-bar-chart", "figure"),
    Input("category-selector", "value")
)
def update_count_orderbar_chart(category):
    # Group data by the selected category
    
    #category="store location"
    res=sales_weather.aggregate([
        {"$group": {"_id": f"${category}", "NbrofOrders": {"$sum": 1}}},
        ])
    filtered_data = pd.DataFrame(res).reset_index();
    print(filtered_data)
    # Create bar chart
    fig = px.bar(filtered_data, x="_id", y="NbrofOrders", title=f"Sales Amount by {category}", 
             labels={"NbrofOrders": "Number of Orders", "_id": category},
             color_discrete_sequence=["skyblue"])

    # Show the figure
    fig.show()
    return fig    
```

```python
# Callback to update the bar chart-sales
@app.callback(
    Output("weather-line-chart", "figure"),
    Input("store-location", "value"),
)
def update_line_chart(val_store_location):
    # Group data by the selected category

    print(val_store_location)
    res2=sales_weather.find(
       { "store_location": val_store_location} 
     )
    print(res2)

    filtered_data= pd.DataFrame(list(res2))
    #filtered_data = pd.DataFrame(res2).reset_index();
    print(filtered_data)
    
    # Create line chart
    fig2 = px.line(filtered_data, x="date", y="Temperature_celsius", title="Temperature Trend by Date", 
              labels={"temperature": "Temperature (Â°C)", "date": "Date"},
              line_shape="linear", color_discrete_sequence=["skyblue"])
    
   
    # Show the figure
    fig2.show()
    return fig2
```

```python
# Callback to update the bar chart-sales
@app.callback(
    Output("weather-bar-chart", "figure"),
    Input("store-location", "value"),
)
def update_weather_bar_chart(val_store_location):
    # Group data by the selected category

    print(val_store_location)
    res2=sales_weather.find(
       { "store_location": val_store_location} 
     )
    print(res2)

    filtered_data= pd.DataFrame(list(res2))
    #filtered_data = pd.DataFrame(res2).reset_index();
    print(filtered_data)
    
    melted_data = filtered_data.melt(value_vars=["Temperature_celsius", "Humidity_perc"], 
                                     var_name="Temperature,Humidity", value_name="Value")
    # Create line chart
  # Create bar chart
    fig3 = px.bar(melted_data, 
                 x="Temperature,Humidity", 
                 y="Value", 
                 title="Temperature & Humidity Values",
                 color="Temperature,Humidity",
                 color_discrete_map={"Temperature_celsius": "skyblue", "Humidity_perc": "orange"})
   
   
    # Show the figure
    fig3.show()
    return fig3
```
